```python
import matplotlib.pylab as plt
import numpy as np
import glob
import os

folderpath = '3m_3k'
filepath = '3m_3k/traces*.bin'

list_of_files = glob.glob(filepath) 
latest_file = max(list_of_files, key=os.path.getctime)
print('latest file is:',latest_file)

list = os.listdir(folderpath) 
number_files = len(list)
print('total file number is:',number_files)

one_trace = np.fromfile(latest_file, dtype=np.float64)
plt.plot(one_trace)
plt.title("Power plot for %s" %(latest_file))
```
