```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fp = "cap_list_byte 0.txt"
F1 = open(fp,'r')
file_string = F1.read()
F1.close()
file_split = file_string.strip().split("\n")
```
