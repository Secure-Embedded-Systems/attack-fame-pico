# plot Samples vs. Max Peak Correlation for all 16 subkeys

```python
from tqdm.notebook import trange
import numpy as np
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt

full_trace = np.fromfile("full_power.bin", dtype=np.float64)
g100_trace = np.fromfile("g100_power.bin", dtype=np.float64)
g200_trace = np.fromfile("g200_power.bin", dtype=np.float64)
g400_trace = np.fromfile("g400_power.bin", dtype=np.float64)
g800_trace = np.fromfile("g800_power.bin", dtype=np.float64)
```

```python
from matplotlib.pyplot import figure
figure(figsize=(14, 8), dpi=80)
plt.plot(full_trace,label='full')
plt.plot(g100_trace,label='g100')
plt.plot(g200_trace,label='g100')
plt.plot(g400_trace,label='g400')
plt.plot(g800_trace,label='g800')
plt.legend()
plt.title('COAES PICO JOULES Power Plot with N-gate removed')
plt.xlabel('Samples')
plt.ylabel('Power')
plt.savefig('trace.jpg',dpi=300) # save jpg/png
```
